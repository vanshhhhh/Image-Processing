```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

import cv2
```

### Loading Image

```python
pil = Image.open("assets/images/Eminem1.jpg")
pil
```

### Shape function gives the number of rows, columns and depth respectively

```python
arr = np.array(pil)
arr.shape
```

### Showing only red layer of the image

```python
red = arr.copy()
red[:,:,[1,2]] = 0
plt.imshow(red)
```

### Showing only green layer of the image

```python
green = arr.copy()
green[:,:,[0,2]] = 0
plt.imshow(green)
```

### Showing only blue layer of the image

```python
blue = arr.copy()
blue[:,:,[0,1]] = 0
plt.imshow(blue)
```

### Converting the image into B&W

```python
gray = arr.mean(axis=2)
plt.imshow(gray, cmap = "gray")
```

## Using OpenCV

```python
frame = cv2.imread("assets/images/Eminem2.jpg")
plt.imshow(frame)
```

#### OpenCV uses BGR instead of RGB that's why the image have different colour

### Converting BGR to RGB for OpenCV

```python
rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

### Focusing on face

#### Zooming on the face using haarcascade

```python
classifier = cv2.CascadeClassifier("assets/data/haarcascade_frontalface_default.xml")
faces = classifier.detectMultiScale(frame)
face = faces[0]
x,y,w,h= face
face = rgb[y:y+h+50,x:x+w+50]
plt.imshow(face)
```

#### Drawing Rectangle on the original image

```python
rectangle = cv2.rectangle(rgb,(x,y),(x+w,y+h),(0,0,225),10)
plt.imshow(rectangle)
```

##### Here the parameters are loaded image, starting coordinate of rectangle, ending coordinate of rectangle, colour of rectangle and thickness of rectangle
