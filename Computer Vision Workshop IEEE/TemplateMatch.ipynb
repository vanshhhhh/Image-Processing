```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('data/image.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
```

```python
gray_image = cv2.imread('data/image.png',0)
plt.imshow(gray_image,cmap="gray")
```

```python
person = cv2.imread('data/person.png')
person = cv2.cvtColor(person, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(person)
```

```python
gray_person = cv2.imread('data/person.png',0)
plt.imshow(gray_person,cmap="gray")
```

```python
result = cv2.matchTemplate(gray_image, gray_person, cv2.TM_CCOEFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
```

```python
top_left = max_loc
bottom_right = (top_left[0] + 200, top_left[1] + 200)
cv2.rectangle(image, top_left, bottom_right, (0,0,255), 5)
plt.imshow(image)
```
